### scikit-learn에서 제공하는 데이터셋 확인
* load : 내장된 데이터셋 불러옴
* fetch : 인터넷을 통해 내려받는 대량 데이터
* make : 확률분포에 근거해서 생성하는 가상 데이터

```python
from sklearn.datasets import load_boston
from sklearn.datasets import load_iris
from sklearn.datasets import load_digits
```

```python
# 데이터 셋 불러오기
iris = load_iris()
boston = load_boston()
digits = load_digits()
```

```python
# 보스턴 집값 데이터 불러오기 2
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
# 보스턴 집값 데이터 대안 1 - 캘리포니아 집값 데이터
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
# 보스턴 집값 데이터 대안 2 - 주택가격 데이터
from sklearn.datasets import fetch_openml
housing = fetch_openml(name="house_prices", as_frame=True)
```

### 인터넷으로 내려받은 데이터셋 확인


```python
import PIL
from sklearn.datasets import fetch_lfw_people
```

```python
people = fetch_lfw_people(min_faces_per_person=25,resize=0.4)
```

```python
print(people.DESCR)
```

```python
people.images[:1]
```

```python
people.target[:1]
```

```python
people.target_names[10]
```

```python
imglbl = list(zip(people.images,digits.target))

plt.figure(figsize=(20,15))
for ix, (img, lbl) in enumerate(imglbl[:50]):
    plt.subplot(5,10,ix+1)
    plt.axis(False)
    plt.imshow(img, interpolation='nearest',cmap=plt.cm.gray)
    plt.title(f'{people.target_names[lbl]}')    
```

```python
people2 = fetch_lfw_people(min_faces_per_person=25,resize=0.4, color=True)
```

```python
imglbl = list(zip(people2.images,people2.target))

plt.figure(figsize=(20,15))
for ix, (img, lbl) in enumerate(imglbl[:50]):
    plt.subplot(5,10,ix+1)
    plt.axis(False)
    plt.imshow(img, interpolation='nearest')
    plt.title(f'{people2.target_names[lbl]}')    
```
